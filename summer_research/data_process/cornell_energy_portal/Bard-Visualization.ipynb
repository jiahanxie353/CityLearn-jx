```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Example for plotting
x = [1,2,3]
y = [n**2 for n in x]
z = [n**3 for n in x]
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(['this is y', 'this is z'])
plt.show()
```

```python
# Load data from CSV files
bard_data = pd.read_csv('Bard Data June 2016 - May 2022.csv')
bard_data
```

```python
filt_chilled_water = (bard_data['pointTitle'] == 'Chilled Water')
```

```python
frame_chilled_water = bard_data.loc[filt_chilled_water]
```

```python
frame_chilled_water
```

```python
slottime = frame_chilled_water.slottime_GMT
slotavg = frame_chilled_water.slotavg
plt.title('Bard Hall Chilled Water Plot')
plt.xlabel('Time/year')
plt.ylabel('Chilled water flow/ton of refrigeration')
plt.plot(slottime, slotavg)
plt.xticks(['2016-06-02 00:00:00','2017-06-01 00:00:00','2018-06-01 00:00:00','2019-06-01 00:00:00','2020-06-01 00:00:00','2021-06-01 00:00:00', '2022-05-01 00:00:00'],['2016','2017','2018','2019','2020','2021','2022'])
```

```python
filt_electric = (bard_data['pointTitle'] == 'Electric')
frame_electric = bard_data.loc[filt_electric]
slottime = frame_electric.slottime_GMT
slotavg = frame_electric.slotavg
plt.title('Bard Hall Electric Plot')
plt.xlabel('Time/year')
plt.ylabel('Electric Demand/kW')
plt.plot(slottime, slotavg, c = 'green')
plt.xticks(['2016-06-02 00:00:00','2017-06-01 00:00:00','2018-06-01 00:00:00','2019-06-01 00:00:00','2020-06-01 00:00:00','2021-06-01 00:00:00', '2022-05-01 00:00:00'],['2016','2017','2018','2019','2020','2021','2022'])
```

```python
filt_steam = (bard_data['pointTitle'] == 'Steam')
frame_steam = bard_data.loc[filt_steam]
slottime = frame_steam.slottime_GMT
slotavg = frame_steam.slotavg
plt.title('Bard Hall Steam Plot')
plt.xlabel('Time/year')
plt.ylabel('Steam flow/(Lbs/Hr)')
plt.plot(slottime, slotavg, c = 'orange')
plt.xticks(['2016-06-03 00:00:00','2017-06-01 00:00:00','2018-06-01 00:00:00','2019-06-01 00:00:00','2020-06-01 00:00:00','2021-06-01 00:00:00', '2022-05-01 00:00:00'],['2016','2017','2018','2019','2020','2021','2022'])
```
