```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
duffield_data = pd.read_csv('Duffield Data June 2016 - May 2022.csv')
```

```python
filt_chilled_water = (duffield_data['pointTitle'] == 'Chilled Water')
frame_chilled_water = duffield_data.loc[filt_chilled_water]
slottime = frame_chilled_water.slottime_GMT
slotavg = frame_chilled_water.slotavg
plt.title('Duffield Hall Chilled Water Plot')
plt.xlabel('Time/year')
plt.ylabel('Chilled water flow/ton of refrigeration')
plt.plot(slottime, slotavg)
plt.xticks(['2016-06-02 00:00:00','2017-06-01 00:00:00','2018-06-01 00:00:00','2019-06-01 00:00:00','2020-06-01 00:00:00','2021-06-01 00:00:00', '2022-05-01 00:00:00'],['2016','2017','2018','2019','2020','2021','2022'])
```

```python
filt_electric = (duffield_data['pointTitle'] == 'Electric')
frame_electric = duffield_data.loc[filt_electric]
slottime = frame_electric.slottime_GMT
slotavg = frame_electric.slotavg
plt.title('Duffield Hall Electric Plot')
plt.xlabel('Time/year')
plt.ylabel('Electric Demand/kW')
plt.plot(slottime, slotavg, c='green')
plt.xticks(
    ['2016-06-02 00:00:00', '2017-06-01 00:00:00', '2018-06-01 00:00:00', '2019-06-01 00:00:00', '2020-06-01 00:00:00',
     '2021-06-01 00:00:00', '2022-05-01 00:00:00'], ['2016', '2017', '2018', '2019', '2020', '2021', '2022'])
```

```python
filt_electric_rm_outliers = frame_electric.slotavg <= 800
sub_frame_electric = frame_electric.loc[filt_electric_rm_outliers]
sub_frame_electric
```
