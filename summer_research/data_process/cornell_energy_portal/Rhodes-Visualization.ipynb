```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
rhodes_data = pd.read_csv('Rhodes Data Jan 2018 - May 2022.csv')
```

```python
filt_chilled_water = (rhodes_data['pointTitle'] == 'Chilled Water')
frame_chilled_water = rhodes_data.loc[filt_chilled_water]
slottime = frame_chilled_water.slottime_GMT
slotavg = frame_chilled_water.slotavg
plt.title('Rhodes Hall Chilled Water Plot')
plt.xlabel('Time/year')
plt.ylabel('Chilled water flow/ton of refrigeration')
plt.plot(slottime, slotavg)
plt.xticks(['2018-01-02 00:00:00','2019-01-01 00:00:00','2020-01-01 00:00:00','2021-01-01 00:00:00', '2022-01-01 00:00:00'],['2018','2019','2020','2021','2022'])
```

```python
filt_electric_1 = (rhodes_data['pointTitle'] == 'Electric meter A')
filt_electric_2 = (rhodes_data['pointTitle'] == 'Electric meter B')
frame_electric = rhodes_data.loc[filt_electric_1]
slottime = frame_electric.slottime_GMT
slotavg = frame_electric.slotavg
plt.title('Hollister Hall Electric Plot')
plt.xlabel('Time/year')
plt.ylabel('Electric Demand/kW')
plt.plot(slottime, slotavg, c='green')
plt.xticks(['2018-01-02 00:00:00','2019-01-01 00:00:00','2020-01-01 00:00:00','2021-01-01 00:00:00', '2022-01-01 00:00:00'],['2018','2019','2020','2021','2022'])
```
