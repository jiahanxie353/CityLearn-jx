```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sage_data = pd.read_csv('Sage Data Jan 2018 - June 2022.csv')
```

```python
filt_chilled_water = (sage_data['pointTitle'] == 'Chilled Water')
frame_chilled_water = sage_data.loc[filt_chilled_water]
slottime = frame_chilled_water.slottime_GMT
slotavg = frame_chilled_water.slotavg
plt.title('Sage Hall Chilled Water Plot')
plt.xlabel('Time/year')
plt.ylabel('Chilled water flow/ton of refrigeration')
plt.plot(slottime, slotavg)
plt.xticks(['2018-01-01 00:00:00','2019-01-01 00:00:00','2020-01-01 00:00:00','2021-01-01 00:00:00', '2022-01-01 00:00:00'],['2018','2019','2020','2021','2022'])
```

```python
filt_cw_rm_outliers = frame_chilled_water.slotavg > 200
sub_frame_cw = frame_chilled_water.loc[filt_cw_rm_outliers]
sub_frame_cw
```

```python
filt_electric = (sage_data['pointTitle'] == 'Electric')
frame_electric = sage_data.loc[filt_electric]
slottime = frame_electric.slottime_GMT
slotavg = frame_electric.slotavg
plt.title('Sage Hall Electric Plot')
plt.xlabel('Time/year')
plt.ylabel('Electric Demand/kW')
plt.plot(slottime, slotavg, c='green')
plt.xticks(['2018-01-01 00:00:00','2019-01-01 00:00:00','2020-01-01 00:00:00','2021-01-01 00:00:00', '2022-01-01 00:00:00'],['2018','2019','2020','2021','2022'])
```
